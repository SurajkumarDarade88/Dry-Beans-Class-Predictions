Detecting The Outliers In The Data

```python
plt.figure(figsize=(5,10))

for col in data.select_dtypes(include = ['number']).columns:
    plt.subplot(8,2, data.columns.get_loc(col) + 1)  # Adjust the subplot layout as needed
    sns.boxplot(x=data[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

From the above cell we can conclude that almost every column in the dataset contains outliers 

```python
## Applying the IQR method to detect outliers ##
q1= data_numeric.quantile(0.25)
q3= data_numeric.quantile(0.75)
## Calculating the IQR ##
iqr = q3 - q1
## Determining the lower and upper bounds ##
lower_bound = q1-1.5*iqr

upper_bound = q3+1.5*iqr
## Identifying outliers
outliers = data_numeric.apply(lambda x: (x < lower_bound[x.name]) | (x > upper_bound[x.name]))
## counting outliers for each column ##
outlier_count = outliers.sum()
```

```python
outlier_count
```

Using capping method to treat the outliers as outlier numbers are higher and convexarea parameter has major impact on the bean quality so we cannot neglect it

```python
def cap_outliers(series, lower_percentile=1, upper_percentile=99):
    lower_threshold = series.quantile(lower_percentile / 100)
    upper_threshold = series.quantile(upper_percentile / 100)
    
    # Cap the values below the lower threshold and upper threshold
    capped_series = series.clip(lower=lower_threshold, upper=upper_threshold)
    return capped_series
```

```python
df=data_numeric.apply(cap_outliers)
df
```

```python
skew = df.skew()
skew
```

To deal with the overfitting we have to do hyperparameter tunning using GridsearchCV

```python
# Hyperparameter tuning using GridSearchCV #
param_grid = {'max_depth': list(range(1,20)),'min_samples_split': [2,5,10],'min_samples_leaf':[1,2,4]}
grid_search = GridSearchCV(estimator=classifier_dt, param_grid=param_grid, cv=5, n_jobs=-1, scoring='accuracy')
grid_search.fit(X_train, Y_train)
```

```python
# Creating the variable for finding the best model #
best_dt = grid_search.best_estimator_
```

```python
# Calculating the Training accuracy #
train_accuracy = best_dt.score(X_train, Y_train)
train_accuracy
```

```python
# Predict the test set results #
Y_pred = best_dt.predict(X_test)
Y_pred
```
